```python
import pandas as pd #to work on dataset
import matplotlib.pyplot as plt #to plot graphs
import collections as co #to get access to containers
df=pd.read_csv("summer.csv")
df.head()
```

**1.In how many cities Summer Olympics is held so far?**

```python
print(len(df['City'].unique())) #no of cities
print(df['City'].unique()) #list of cities
```

**2.Which sport is having most number of Gold Medals so far? (Top 5)**

```python
data = []
for i in range(0,len(df)):
  if(df['Medal'][i]=='Gold'):
    data.append(df['Sport'][i]) #corresponding sport with gold medal is added to the list
```

```python
data #displayed and verified
```

```python
x=co.Counter(data)
x.most_common(5) #top 5 sports with gold medals
```

**3.Which sport is having most number of medals so far? (Top 5)**

```python
data = []
for s in df['Sport'].unique():
    data.append([s , len(df[df['Sport']  == s])])
pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (15,5))
#top 5 could be identified from graph
```

**4.Which player has won most number of medals? (Top 5)**

```python
l=[]
for i in range(0,len(df)):
  l.append(df['Athlete'][i])
x=co.Counter(l)
x.most_common(5) #top 5 players with most medals
```

**5.Which player has won most number Gold Medals of medals?(Top 5)**

```python
data = []
for i in range(0,len(df)):
  if(df['Medal'][i]=='Gold'):
    data.append(df['Athlete'][i]) #corresponding value under Athlete column is added to the list
x=co.Counter(data)
x.most_common(5) #top 5 athletes are viewed
```

**6.In which year India won first Gold Medal in Summer Olympics?**

```python
l=[]
for i in range(len(df)):
  if(df['Country'][i]=='IND' and df['Medal'][i]=='Gold'):
    l.append(df['Year'][i]) #all the years during which india won a gold medal is added to the list
print(l[0]) #the year in which india won first gold medal is printed
```

**7.Which event is most popular in terms on number of players?(Top 5)**

```python
e=[]
for i in range(0,len(df)):
  e.append(df['Event'][i])
x=co.Counter(e)
x.most_common(5) #popular events along with no of participants is printed
```

**8.Which sport is having most female Gold Medalists?(Top 5)**

```python
w=[]
for i in range(0,len(df)):
  if(df['Gender'][i]=='Women' and df['Medal'][i]=='Gold'):
    w.append(df['Sport'][i])
x=co.Counter(w)
x.most_common(5) #top 5 sports with most female gold medalists
```
